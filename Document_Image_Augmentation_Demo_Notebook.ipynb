## Code API

````
# Check your working directory, change the import path accordingly 
%cd "/content/Document-Image-Augmentation"
import os
import argparse
from google.colab.patches import cv2_imshow 
import cv2
from DocumentAugmentation import DocumentAugmentation
import glob 
````

````
# Read the image
src = "demo.jpg"
image = cv2.imread(src)
# Change this to cv2.imshow() when not using colab, i.e if you are running locally
cv2_imshow(image) 
````

### Perform Augmentation

````
Dialated = DocumentAugmentation.Dialate(image) # Do not use for OCR
Smudged = DocumentAugmentation.Smudge(image) # Do not use for OCR 
colored = DocumentAugmentation.changeColor(image) # this will randomly select color
BUimage = DocumentAugmentation.BrightnessUp(image)
BDimage = DocumentAugmentation.BrightnessDown(image)
````

### Results

````
cv2_imshow(Dialated)
cv2_imshow(Smudged)
cv2_imshow(colored)
cv2_imshow(BUimage)
cv2_imshow(BDimage)
````
